# Lecture 3. Summary functions and Maps

```python
wines_filepath = 'winemag-data-130k-v2.csv'
reviews = pd.read_csv(wines_filepath, index_col=0)
reviews
```

```python
reviews.points.describe()
```

```python
reviews.taster_name.describe()
```

```python
reviews.points.mean()
```

```python
reviews.taster_name.unique()
```

```python
reviews.taster_name.value_counts()
```

```python
reviews_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - reviews_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - reviews_points_mean
    return row
reviews.apply(remean_points, axis='columns')
```

```python
review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
```

```python
reviews.country + "-" + reviews.region_1
```

# Lecture 4. Grouping and sorting

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
countries_reviewed.reset_index()
```

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
countries_reviewed.sort_index()
```

```python
countries_reviewed.sort_values(by=['country', 'len'])
```
